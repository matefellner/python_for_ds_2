Using functions in slicing

```python
def rating_function(x):
    if x >= 8.0:
        return "good"
    else:
        return "bad"

df["rating_category"] = df["rating"].apply(rating_function)

df.head(2)
```

## Grouping, aggregation and joining

**`value_counts`** counts how often each value occurs:

```python
df['rating_category'].value_counts()
```

**`sort_values`** orders the rows by a column:

```python
# The 5 highest-rated movies
df.sort_values('rating', ascending=False)[['rating', 'director', 'year']].head()
```

**`groupby`** splits the rows by a key, then aggregates each group:

```python
# Average rating per year
df.groupby('year')['rating'].mean()
```

```python
# Several aggregates at once, per rating category
df.groupby('rating_category')[['rating', 'revenue', 'votes']].agg(['mean', 'count'])
```

**`merge`** joins two tables on a shared column (like a SQL join):

```python
# A second table keyed by director
directors = pd.DataFrame({'director': ['Christopher Nolan', 'Ridley Scott', 'James Gunn'],
                          'country': ['UK', 'UK', 'US']})

# Attach each director's country to their movies by matching on 'director'
df[['director', 'rating']].head().merge(directors, on='director', how='left')
```

### Data cleaning

Dropping columns

```python
df = pd.read_csv('data/BL-Flickr-Images-Book.csv')
df.head()
```

```python
to_drop = ['Edition Statement',
           'Corporate Author',
           'Corporate Contributors',
           'Former owner',
           'Engraver',
           'Contributors',
           'Issuance type',
           'Shelfmarks']

df.drop(to_drop, inplace=True, axis=1)
#df.drop(columns=to_drop, inplace=True)
```

```python
df.head()
```

Changing index

```python
df['Identifier'].is_unique
df = df.set_index('Identifier')
df.head()
```

Field tidying

```python
extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype
```

Data cleaning with string methods and NumPy

```python
df.loc[4157862]
```

```python
df.loc[4159587]
```

```python
pub = df['Place of Publication']
london = pub.str.contains('London', na=False)
print(london[:5])
oxford = pub.str.contains('Oxford', na=False)
```

```python
df['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))

df['Place of Publication'].head()
```
